```python
import numpy as np
import cv2 as cv
import os
```

```python
minDist = 10
param1 = 100 #500
param2 = 10 #200 #smaller value-> more false circles
minRadius = 5
maxRadius = 8 #10
dp = 1

blur_kernel = 5

game_index = 1

circle_threshold = 0.5
```

```python
def find_circles(img):
    image_circle = img.copy()
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    blurred = cv.GaussianBlur(gray, (blur_kernel, blur_kernel), 0)
    circles = cv.HoughCircles(blurred, cv.HOUGH_GRADIENT, dp, minDist, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)

    valid_circles = []
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        for (x, y, r) in circles:
            # count the number of black pixles in the circle
            black_count = 0
            cnt_pixels = 0
            for i in range(x-r, x+r):
                for j in range(y-r, y+r):
                    if (i-x)**2 + (j-y)**2 <= r**2 and i >= 0 and j >= 0 and i < img.shape[1] and j < img.shape[0]:
                        if gray[j][i] == 0:
                            black_count += 1
                        cnt_pixels += 1

            if black_count/cnt_pixels > circle_threshold:
                # cv.circle(image_circle, (x, y), r, (0, 0, 255), 2)
                valid_circles.append((x, y, r))
    return valid_circles
```

```python
images_path = "../../extracted_domino_hq_clean/"
for file in sorted(os.listdir(images_path)):
    if file.endswith(".jpg") and file.startswith(str(game_index) + "_"):
        cnt_0 = 0
        cnt_1 = 0
        img = cv.imread(images_path + file)
        print(type(img), img.shape)
        circles = find_circles(img)

        img_circle = img.copy()
        for circle in circles:
            cv.circle(img_circle, (circle[0], circle[1]), circle[2], (0, 0, 255), 2)

        print(img.shape)
        print(f"There are {len(circles)}: {circles}")
        if img.shape[0] < img.shape[1]:
            # draw a vertical line in the middle
            cv.line(img_circle, (img.shape[1]//2, 0), (img.shape[1]//2, img.shape[0]), (0, 255, 0), 2)

            for circle in circles:
                if circle[0] < img.shape[1]//2:
                    cnt_0 += 1
                else:
                    cnt_1 += 1

        else:
            # draw a horizontal line in the middle
            cv.line(img_circle, (0, img.shape[0]//2), (img.shape[1], img.shape[0]//2), (0, 255, 0), 2)

            for circle in circles:
                if circle[1] < img.shape[0]//2:
                    cnt_0 += 1
                else:
                    cnt_1 += 1
        
        cv.imshow("circles", img_circle)
        cv.waitKey(0)
        print("cnt_0: ", cnt_0)
        print("cnt_1: ", cnt_1)
        break

cv.destroyAllWindows()
```

```python
# Load image, grayscale, median blur, Otsus threshold
image = cv.imread('../../extracted_domino_pieces/1_01.jpg')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
blur = cv.medianBlur(gray, 11)
thresh = cv.threshold(blur, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)[1]

# Morph open 
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5,5))
opening = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel, iterations=3)

# Find contours and filter using contour area and aspect ratio
cnts = cv.findContours(opening, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
for c in cnts:
    peri = cv.arcLength(c, True)
    approx = cv.approxPolyDP(c, 0.04 * peri, True)
    area = cv.contourArea(c)
    if len(approx) > 5 and area > 1000 and area < 500000:
        ((x, y), r) = cv.minEnclosingCircle(c)
        cv.circle(image, (int(x), int(y)), int(r), (36, 255, 12), 2)

cv.imshow('thresh', thresh)
cv.imshow('opening', opening)
cv.imshow('image', image)
cv.waitKey()     
cv.destroyAllWindows()
```
